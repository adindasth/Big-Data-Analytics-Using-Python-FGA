# Pengolahan data

```python
#Mengimpor library yang akan digunakan 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Melakukan read data csv menggunakan pandas
d_parser = lambda x: pd.datetime.strptime(x, '%d/%m/%Y')
df = pd.read_csv('iklim.csv', parse_dates=['Tanggal'], date_parser=d_parser)
df.head(50)
```

```python
#Mengubah data yang tidak valid dengan NaN
df = df.replace([8888.0, 9999.0], np.NaN)
df.head(50)
```

```python
#Mengisi data NaN dengan data sebelumnya
df.fillna(method='ffill',inplace=True)
df.head(50)
```

```python
#Menghapus kolom Temperatur mininum, Temperatur maksimum, dan Arah angin saat kecepatan maksimum
df.drop(['Tn', 'Tx', 'ddd_x'], axis=1, inplace=True)
df.info()
```

```python
df
```

# Visualisasi

```python
#Visualisasi
# Melakukan visualisasi untuk kolom Tavg
plt.figure(figsize = (20, 8))
plt.title("Visualisasi Data Untuk Setiap Tavg", fontsize=10)
sns.countplot(df['Tavg'])
plt.show()

# Melakukan visualisasi untuk kolom RH_avg
plt.figure(figsize = (10, 5))
plt.title("Visualisasi Data Untuk Setiap RH_avg", fontsize=10)
sns.countplot(df['RH_avg'])
plt.show()

# Melakukan visualisasi untuk kolom RR
plt.figure(figsize = (25, 8))
plt.title("Visualisasi Data Untuk Setiap RR", fontsize=10)
sns.countplot(df['RR'])
plt.show()

# Melakukan visualisasi untuk kolom ss
plt.figure(figsize = (20, 8))
plt.title("Visualisasi Data Untuk Setiap ss", fontsize=10)
sns.countplot(df['ss'])
plt.show()

# Melakukan visualisasi untuk kolom ff_x
plt.figure(figsize = (8, 5))
plt.title("Visualisasi Data Untuk Setiap ff_x", fontsize=10)
sns.countplot(df['ff_x'])
plt.show()

# Melakukan visualisasi untuk kolom ff_avg
plt.figure(figsize = (8, 5))
plt.title("Visualisasi Data Untuk Setiap ff_avg", fontsize=10)
sns.countplot(df['ff_avg'])
plt.show()

# Melakukan visualisasi untuk kolom ddd_car
plt.figure(figsize = (8, 5))
plt.title("Visualisasi Data Untuk Setiap ddd_car", fontsize=10)
sns.countplot(df['ddd_car'])
plt.show()
```
