```python
judul = "Konversi Mata Uang"
print(judul.center(60," "))


mata_uang = {'IDR': 1, 'USD': 14479.35, 'GBP': 20020.81, 'SAR': 3860.66, 'JPY': 131.43, 'CNY' : 2233.80}

def konversi (mata_uang_asal, mata_uang_tujuan, masukan):
    return(mata_uang[mata_uang_asal]/mata_uang[mata_uang_tujuan] * masukan)

mata_uang_asal = (input("Pilih mata uang asal anda (IDR, USD, GBP, SAR, JPY, CNY): "))
mata_uang_tujuan = (input("Pilih mata uang tujuan konversi (IDR, USD, GBP, SAR, JPY, CNY): "))
masukan = float(input("Masukkan jumlah mata uang yang ingin dikonversi: "))

print("Nilai mata uang anda setelah konversi adalah", konversi(mata_uang_asal, mata_uang_tujuan,masukan), mata_uang_tujuan)  
```
