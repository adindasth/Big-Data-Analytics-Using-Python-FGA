```python
from datetime import datetime

class BankAccount:
    
    __MIN_BALANCE = 50000
    
#Pertama buatlah konstruktor dari class Bank berupa:
    #1. nama pemilik akun bank
    #2. kode rekening bank
    #3. tanggal transaksi terakhir
    #4. saldo terakhir
     
    def __init__(self, owner, account_number):
        self.owner = owner
        self.account_number = account_number
        self.last_transaction_date = datetime.now()
        self.last_transaction_date.strftime("%A, %d %B %Y %X")
        self.balance = self.__MIN_BALANCE 
        pass

# Kedua buatlah method untuk menambah uang ke dalam akun pemilik.
    def deposit(self, amount):
        self.balance += amount
        pass

# Ketiga buatlah method untuk melakukan penarikan uang dari akun pemilik.
# Apabila pemilik ingin menarik uang lebih dari nilai saldo + minimal saldo maka transaksi gagal dan saldo tidak berubah.
    def withdraw(self, amount):
        if amount > self.balance:
            return False
        else:
            self.balance -= amount
        pass
        
# Keempat buatlah fungsi _str_ untuk mencetak informasi tentang objek pemilik akun
    def __str__(self):
        print("\tBank Account: ")
        print("\t   Account Owner:", self.owner)
        print("\t   Account Number:", self.account_number)
        print("\t   Last Transaction:", self.last_transaction_date.strftime("%A, %d %B %Y %X"))
        return("\t   Current Balance: {0}".format(self.balance))
        pass
```

```python
adinda = BankAccount("Adinda Siti Shalehah", "123456")
annisa = BankAccount("Annisa Fitri", "456789")
print(adinda)
print(annisa)
```

```python
adinda.deposit(50000)
annisa.deposit(200000)
print(adinda)
print(annisa)
```

```python
adinda.withdraw(150000)
annisa.withdraw(25000)
print(adinda)
print(annisa)
```
